```python
Q1. Pearson correlation coefficient is a measure of the linear relationship between two variables. Suppose
you have collected data on the amount of time students spend studying for an exam and their final exam
scores. Calculate the Pearson correlation coefficient between these two variables and interpret the result.
Ans:
    The Pearson correlation coefficient, denoted by "r," ranges from -1 to 1 and measures the strength and direction of a linear
    relationship between two variables. A positive "r" value indicates a positive correlation, meaning that as one variable
    (study time) increases, the other variable (final exam scores) tends to increase as well. On the other hand, a negative "r" 
    value indicates a negative correlation, where an increase in one variable is associated with a decrease in the other.

If the calculated "r" value is close to 1, it indicates a strong positive linear relationship, implying that students who study 
more tend to achieve higher exam scores. Conversely, if "r" is close to -1, it suggests a strong negative linear relationship, 
implying that students who spend less time studying may achieve higher exam scores.

However, if the "r" value is close to 0, it indicates a weak or no linear relationship between study time and exam scores. In 
this case, the two variables may not have a significant linear association, and other factors could be influencing students'
final exam performance.
```

```python
Q2. Spearman's rank correlation is a measure of the monotonic relationship between two variables.
Suppose you have collected data on the amount of sleep individuals get each night and their overall job
satisfaction level on a scale of 1 to 10. Calculate the Spearman's rank correlation between these two
variables and interpret the result.
Ans:
    Spearman's rank correlation, denoted by "ρ" (rho), is a measure of the strength and direction of the monotonic relationship
    between two variables. Unlike Pearson correlation, Spearman's correlation assesses the relationship based on the ranks of 
    the data rather than the actual values.

To calculate Spearman's rank correlation for the given data on sleep and job satisfaction, follow these steps:

1. Rank the data for each variable separately (e.g., rank the sleep values and rank the job satisfaction scores).
2. Calculate the difference in ranks for each data point between the two variables.
3. Square these differences and sum them up.
4. Use the formula for Spearman's rank correlation to calculate the coefficient.

Interpretation:
- If the Spearman's rank correlation coefficient (ρ) is close to +1, it indicates a strong positive monotonic relationship. This
means that as the amount of sleep individuals get each night increases, their overall job satisfaction tends to increase as well
.
- If ρ is close to -1, it implies a strong negative monotonic relationship. In this case, as the amount of sleep decreases, job
satisfaction tends to decrease.
- If ρ is close to 0, it suggests a weak or no monotonic relationship between the two variables. This means that changes in 
sleep do not show a consistent pattern in relation to job satisfaction, or there might be other factors influencing job satisfa
ction levels.

Keep in mind that Spearman's rank correlation only captures monotonic relationships and does not imply causation. Other factors 
could also contribute to job satisfaction levels beyond the amount of sleep individuals get.
```

```python
Q3. Suppose you are conducting a study to examine the relationship between the number of hours of
exercise per week and body mass index (BMI) in a sample of adults. You collected data on both variables
for 50 participants. Calculate the Pearson correlation coefficient and the Spearman's rank correlation
between these two variables and compare the results.
Ans:
    To calculate the Pearson correlation coefficient and Spearman's rank correlation for the given data, follow these steps:

1. Compute the Pearson correlation coefficient (r):
   - Calculate the mean of the number of hours of exercise per week (let's call it "X̄") and the mean of BMI (let's call it "Ȳ")
.
   - Calculate the standard deviation of the number of hours of exercise per week (let's call it "σX") and the standard deviat-
                                                                                   ion of BMI (let's call it "σY").
   - For each participant, compute the z-scores for the number of hours of exercise (Zx = (X - X̄) / σX) and the z-scores for BMI
                                                                                               (Zy = (Y - Ȳ) / σY).
   - Multiply the z-scores for each participant and sum them up (Σ(Zx * Zy)).
   - Divide the sum by (n - 1), where "n" is the number of participants, to get the Pearson correlation coefficient (r).

2. Compute Spearman's rank correlation coefficient (ρ):
   - Rank the data for both variables separately (rank the number of hours of exercise and rank the BMI values).
   - Calculate the difference in ranks for each data point between the two variables.
   - Square these rank differences and sum them up.
   - Use the formula for Spearman's rank correlation to calculate the coefficient:
     ρ = 1 - (6 * Σ(d^2)) / (n * (n^2 - 1)), where "d" is the rank difference for each data point, and "n" is the number of par-
                                                                                               ticipants.

Comparison of results:
- If both the Pearson correlation coefficient (r) and Spearman's rank correlation coefficient (ρ) are close to +1 or -1, it ind-
  icates a strong linear or monotonic relationship, respectively, between the two variables (hours of exercise and BMI). This 
  means that as the number of hours of exercise per week increases, the BMI tends to either decrease or increase in a predicta-
ble pattern.
- If both coefficients are close to 0, it suggests a weak or no linear or monotonic relationship between the two variables. This
  means that there is no consistent pattern or association between hours of exercise and BMI in the sample of adults.

Keep in mind that the interpretation of the results may vary depending on the actual values of "r" and "ρ" obtained from the 
data.
```

```python
To calculate the Pearson correlation coefficient (r) between the number of hours individuals spend watching television per day
and their level of physical activity, follow these steps:

1. Compute the mean of the number of hours individuals spend watching television per day (let's call it "X̄") and the mean of 
    their level of physical activity (let's call it "Ȳ").

2. Calculate the standard deviation of the number of hours individuals spend watching television per day (let's call it "σX")
    and the standard deviation of their level of physical activity (let's call it "σY").

3. For each participant, compute the z-scores for the number of hours of television watching (Zx = (X - X̄) / σX) and the z-scor-
    es for the level of physical activity (Zy = (Y - Ȳ) / σY).

4. Multiply the z-scores for each participant and sum them up (Σ(Zx * Zy)).

5. Divide the sum by (n - 1), where "n" is the number of participants, to get the Pearson correlation coefficient (r).

The obtained "r" value will indicate the strength and direction of the linear relationship between the number of hours individ-
uals spend watching television and their level of physical activity.

Please note that the specific values for the data are needed to calculate the actual Pearson correlation coefficient, and with-
out the data, it is not possible to provide the exact result. The "r" value can range from -1 to 1, where a positive value ind-
icates a positive correlation (as one variable increases, the other tends to increase), a negative value indicates a negative
correlation (as one variable increases, the other tends to decrease), and a value close to 0 indicates a weak or no linear rela-
tionship between the two variables.
```
